```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt




df = pd.read_csv("/content/housing_in_london_monthly_variables2.csv")
df.head(5)
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.regplot(x="houses_sold", y="average_price", data=df)
```

```python
sns.lmplot(x="houses_sold", y="average_price", data=df,hue='date')
```
