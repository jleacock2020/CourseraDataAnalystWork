```python
import pandas as pd
!pip install pywaffle
import matplotlib.pyplot as plt
from pywaffle import Waffle
```

```python
df = pd.read_csv("/content/Car Features and MSRP.csv")
```

```python
car_make_counts = df['Make'].value_counts()
car_make_data = pd.DataFrame({'Car Make': car_make_counts.index, 'Count': car_make_counts.values})
```

```python
# Get the top 5 car makes based on their counts
top_car_makes = df['Make'].value_counts().nlargest(5)

# Create a new DataFrame for the waffle chart
car_make_data = pd.DataFrame({'Car Make': top_car_makes.index, 'Count': top_car_makes.values})

# Create the waffle chart
fig = plt.figure(
    FigureClass = Waffle,
    rows = 5,
    columns = 18
  ,
    vertical = True,
    values = car_make_data['Count'],
    labels = list(car_make_data['Car Make']),
    legend = {'loc': 'upper left', 'bbox_to_anchor': (1.05, 1), 'fontsize': 12},
    figsize = (10, 6)
)

plt.title("Top 5 Car Makes Distribution")
plt.show()

```
